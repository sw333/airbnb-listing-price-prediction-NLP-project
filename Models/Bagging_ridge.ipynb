```
import pandas as pd
import numpy as np
pathlistings = '/content/gdrive/My Drive/BT4222 project/afterfeatureengineering2.csv'
df_listings = pd.read_csv(pathlistings)
len(df_listings)
```

```
X = df_listings.drop(columns=['price'])
y = df_listings.price

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import Ridge
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import ParameterGrid
grid = ParameterGrid({"max_samples": [0.5, 1.0],
                  "max_features": [0.5, 1.0],
                  "bootstrap": [True, False],
                  "bootstrap_features": [True, False],
                  "n_estimators":[20,30,40,50,60,70,80,100]})
RMSE = 10000000000000000
para_rmse = None
R2 = 1000000000000000000000
para_r2 = None

for params in grid:
    print("------------------CURRENT-------------------------")
    print("Current params:",params)
    bagging = BaggingRegressor(base_estimator=Ridge(),
                      random_state=0,
                      **params).fit(X_train, y_train)
    y_pred_train = bagging.predict(X_train) 
    y_pred_test = bagging.predict(X_test) 
    # RMSE
    rmse_train = mean_squared_error(y_train, y_pred_train, squared=False)
    print("This round training RMSE:",rmse_train)
    rmse_test = mean_squared_error(y_test, y_pred_test, squared=False)
    print("This round testing RMSE:",rmse_test)
    # r2
    r2_train = r2_score(y_train, y_pred_train)
    print("This round training R2:",r2_train)
    r2_test = r2_score(y_test, y_pred_test)
    print("This round testing R2:",r2_test)

    if rmse_test < RMSE:
      RMSE = rmse_test
      para_rmse = params
    print("--------------------RMSE-----------------------")
    print("current lowest RMSE:",RMSE)
    print("current best para:", para_rmse)
  
    if r2_test < R2:
      R2 = r2_test
      para_r2 = params
    print("-------------------R2------------------------")
    print("current lowest R2:",R2)
    print("current best para:", para_r2)
```
