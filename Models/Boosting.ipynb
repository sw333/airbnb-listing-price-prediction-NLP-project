```
import pandas as pd
import numpy as np
pathlistings = '/content/gdrive/My Drive/BT4222 project/afterfeatureengineering2.csv'
df_listings = pd.read_csv(pathlistings)
len(df_listings)
```

```
X = df_listings.drop(columns=['price'])
y = df_listings.price

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```
import xgboost

xgb=xgboost.XGBRegressor(n_estimators=450,max_depth=8,random_state=4)
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
y_train_pred = xgb.predict(X_train)
y_pred = xgb.predict(X_test)
# RMSE
print("training RMSE: ", mean_squared_error(y_train, y_train_pred, squared=False))
print("testing RMSE",mean_squared_error(y_test, y_pred, squared=False))
#r2
print("training R2: ", r2_score(y_train, y_train_pred))
print("testing R2: ", r2_score(y_test, y_pred))
#standard deviation
print(np.std(y_test))
```
